```python

plt.title("Box PLot of age and sex")  
sns.boxplot(x='sex',y='age',data=df)
# The box plot compares age distribution between sexes. It shows the median age, quartiles, and any outliers for both males and females. 
```

```python
plt.title("Histrogram of WeightKG") 
plt.hist(df['weightKg'])
# The histogram shows the distribution of weight in kilograms. It helps to visualize the frequency of different weight ranges in the dataset. 
```

```python
plt.title("Scatter Plot") 
sns.scatterplot(x='age', y='BMI', hue='sex', style='indication', data=df)
# The scatter plot shows the relationship between age and BMI, with points colored by sex and styled by indication (likely representing health status or condition). 
```
